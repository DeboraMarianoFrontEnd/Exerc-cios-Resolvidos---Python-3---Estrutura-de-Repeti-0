```python
print('=' * 30)
print(f'{"BANCO CEV":^30}') #deixa centralizado, o 30 é a distancia que quero do inicio até o meio e do fim até o meio.
print('=' * 30)
valor = int(input('Qual valor voce quer sacar? R$'))
total = valor #armazena o valor total a ser sacado.
ced = 50 #o professor quis começar com a cedula de maior valor que é de 50 reais
totced = 0

while True: #esse loop vai parar quando acabar o dinheiro
  if total >= ced:
    total -= ced #aqui eu pego o valor que o usuario colocou e vou tirando 50 (ced = 50)
    totced += 1 #toda vez que for tirado 50, vai ser contabilizado
  else:
    if totced > 0: #so vai printar se a quantidade de cedula for maior que zero para nao acontecer por exemplo de colocar 2 cedulas de R$50, 2 cedulas de R$20, 0 cedulas de R$10...
      print(f'Total de {totced} cédulas de R${ced}')
    if ced == 50: #Troca a denominação da cédula para a próxima menor (de 50 para 20, de 20 para 10, de 10 para 1).
      ced = 20
    elif ced == 20:
      ced = 10
    elif ced == 10:
      ced = 1
    totced = 0 #reseta totced para começar a contar as novas cédulas.
    if total == 0: #aqui é para indicar que quando acabar meu valor, o meu programa pára.
      break

#Esse if total >= ced vai me dizer quantas vezes eu consigo tirar 50 do valor total que o usuario colocou.

#O else foi colocado para que o programa entenda que se caso nao der mais como tirar 50 do meu valor, ai eu tenho que verificar a cedula atual
#Entao colocamos, se a cedula for == 50, a proxima cedula vai virar a de 20, e se a cedula for a de 20 (if ced == 20), eu diminuo para a cedula
#de 10 (ced = 10), mas se a cedula for a de 10, eu diminuo para a cedula de 1.
```

```python
#como um aluno fez:

saque = int(input('Digite o valor que deseja sacar: R$ '))
n50 = n20 = n10 = n1 = 0
while True:
    if saque >= 50:
        saque -= 50
        n50 += 1
    else:
        if saque >= 20:
            saque -= 20
            n20 += 1
        else:
            if saque >= 10:
                saque -= 10
                n10 += 1
            else:
                if saque >= 1:
                    saque -= 1
                    n1 += 1
    if saque == 0:
        break
print(f'Você receberá {n50} notas de 50, {n20} de 20, {n10} de 10 e {n1} de 1.')
```

```python
cont = 1 # tem como papel contar a quantidade de vezes que o while está sendo executado.
while cont <= 4000:
  print("Olá!")
  cont += 1 #se tirar esse contador, ele vai cair no loop infinito e nao vai parar, por isso precisa dele pra contar.
print("FIM")

#quando o valor de cont passar a ser maior que 4000, a condição de execução deixará de ser verdadeira e, consequentemente, as instruções
#definidas nas linhas 3 e 4 não serão executadas.

#cont += 1 -->
todas as vezes que o programa entrar no laço(loop), a variável contadora terá o seu valor aumentado em 1, isto é, cont += 1 (vale
relembrar que esta instrução poderia ser substituída por cont = cont + 1). Ao fim da execução da linha 4, o fluxo de execução retornará para a linha
2, verificando se o novo valor de cont é menor ou igual a 4000. Sendo verdadeiro, as linhas 3 e 4 serão executadas novamente.

Quando cont passar a ser maior que 4000, o laço não será mais executado e o fluxo do programa será deslocado para a linha 5.

#cont = 1 -->
considerando que cont é inicializada com o valor 1, depois de 4000 vezes, seu valor será 4001, ou seja, a condição que sinaliza para o
programa finalizar o laço.
```

```python
soma = 0  #A cada iteração, soma é utilizada para acumular o somatório dos termos da PA
termo = 1
while termo <= 10:
  soma += termo #soma armazenará o resultado da adição entre o valor que ela já possui e o valor que está armazenado em termo
  termo += 1 #geração do termo da PA, de razão 1
print(soma) #se colocar dentro do while a identacao, a cada iteracao, vai mostrar a soma: 0 +1 = 1, na segunda iteracao: 0+1+2=3 e assim por diante.



# A variável termo é incrementada em uma unidade. Perceba que isso corresponde à geração do termo da PA, ou seja, o novo valor para termo é
#o valor que ele possuía na iteração anterior somado à razão 1: termo += 1.
```
